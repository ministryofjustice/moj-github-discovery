```python
# Parse ISO-8601 strings like '2026-03-05T12:35:16Z' into timezone-aware datetimes (UTC)
df["pushed_at_dt"] = pd.to_datetime(df["pushed_at"], utc=True, errors="coerce")

# "Age" since pushed_at (relative to now)
now_utc = pd.Timestamp.now(tz="UTC")
df["age"] = now_utc - df["pushed_at_dt"]  # timedelta64[ns]

# Handy numeric version of age for stats/plots
df["age_days"] = df["age"] / pd.Timedelta(days=1)

# Quick stats
print("pushed_at range:", df["pushed_at_dt"].min(), "->", df["pushed_at_dt"].max())
print("avg age:", df["age"].mean())
print("avg age (days):", df["age_days"].mean())

# Optional: distribution of events by day
by_day = df["pushed_at_dt"].dt.floor("D").value_counts().sort_index()
display(by_day)
```

```python
df['workflows_names'].value_counts()
```

```python

# Explode comma-separated workflow names into one row per workflow
wf_exploded = (
    df[["full_name", "workflows_names"]]
    .copy()
    .assign(workflow=lambda d: d["workflows_names"].str.split(r",\s*"))
    .explode("workflow")
    .assign(workflow=lambda d: d["workflow"].str.strip())
    .query("workflow != '' and workflow.notna()")
    .reset_index(drop=True)
)

# Per-workflow summary: which repos use it and how many
wf_summary = (
    wf_exploded.groupby("workflow", sort=True)
    .agg(
        repo_count=("full_name", "nunique"),
        repos=("full_name", lambda x: ", ".join(sorted(x.unique()))),
    )
    .sort_values("repo_count", ascending=False)
    .reset_index()
)

print(f"Unique workflow names : {wf_summary['workflow'].nunique()}")
print(f"Total workflow usages : {len(wf_exploded)}")
display(wf_summary)
```

```python
# Repos generated from a template — grouped by template source
template_summary = (
    df[["full_name", "template_source"]]
    .dropna(subset=["template_source"])
    .query("template_source != ''")
    .groupby("template_source", sort=True)
    .agg(
        repo_count=("full_name", "nunique"),
        repos=("full_name", lambda x: ", ".join(sorted(x.unique()))),
    )
    .sort_values("repo_count", ascending=False)
    .reset_index()
)

print(f"Unique template sources : {template_summary['template_source'].nunique()}")
print(f"Repos generated from a template: {template_summary['repo_count'].sum()}")
display(template_summary)
```
